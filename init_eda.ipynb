```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data2_03 = pd.read_csv('~/Desktop/Redback/data2/data2_03/output.csv')
data2_23 = pd.read_csv('~/Desktop/Redback/data2/data2_23/output.csv')
data2_47 = pd.read_csv('~/Desktop/Redback/data2/data2_47/output.csv')
data1_09 = pd.read_csv('~/Desktop/Redback/data1/data1_09/output.csv')
data1_21 = pd.read_csv('~/Desktop/Redback/data1/data1_21/output.csv')
data1_30 = pd.read_csv('~/Desktop/Redback/data1/data1_30/output.csv')
```

```python
#we can use speed + laptime to find out which pit stop collisions happened 
data2_03['acsys.CS.LapTime'].plot()
plt.title("LapTime for car 03 (data2)")
plt.xlabel("Time Series (obs number)")
plt.ylabel("Lap Time")

plt.show()
```

```python
data2_03['acsys.CS.SpeedKMH'].plot()
plt.title("Speed for car 03 (data2)")
plt.xlabel("Time Series (obs number)")
plt.ylabel("Speed (KMH)")

plt.show()
```

```python
f, axis = plt.subplots(2, 3)
f.set_figheight(15)
f.set_figwidth(20)

# For car03 (data2) Laptime 
axis[0, 0].plot(data2_03['acsys.CS.LapTime'])
axis[0, 0].set_title("data2_car03 LapTime")
axis[0, 0].set_ylim(0,250000)

# For car23 (data2) Laptime
axis[0, 1].plot(data2_23['acsys.CS.LapTime'])
axis[0, 1].set_title("data2_car23 LapTime")
axis[0, 1].set_ylim(0,250000)

# For car47 (data2) Laptime
axis[0, 2].plot(data2_47['acsys.CS.LapTime'])
axis[0, 2].set_title("data2_car47 LapTime")
  
# For car03 (data2) SpeedKMH
axis[1, 0].plot(data2_03['acsys.CS.SpeedKMH'])
axis[1, 0].set_title("data2_car03 Speed")
axis[1, 0].set_ylim(0,100)

# For car23 (data2) SpeedKMH
axis[1, 1].plot(data2_23['acsys.CS.SpeedKMH'])
axis[1, 1].set_title("data2_car23 Speed")
axis[1, 1].set_ylim(0,100)

# For car23 (data2) SpeedKMH
axis[1, 2].plot(data2_47['acsys.CS.SpeedKMH'])
axis[1, 2].set_title("data2_car47 Speed")
axis[1, 2].set_ylim(0,100)
  
plt.show()
```
